```python
#condicionando bien la grafica para que cuando llegue a la cesta, caiga
m=[] #actualiza las valores en y menores a la altura de la cesta
x1=[] #actualiza las posciones x hasta la distancia maxima
p=np.where(y==g)[0][0]
s=[] #actualiza los valores en y mayores a la altura de la cesta
for i in range(len(y)):
  if i<p:
    m.append(y[i])
    x1.append(x[i])

  if i>=p:
    s.append(y[p])
    x1.append(x[p])

epsilon=np.linspace(0,3,len(s))

s=np.array(s)
resta= s-epsilon
resta=list(resta)

valores= m + resta #une las listas con los valores actuales adecuados para una trayectoria con sentido fisico
```

```python
plt.figure(figsize=(15,4))
plt.plot(x1,valores, label= 'trayectoria del balon')
plt.plot(7,3,'.',label = 'cesta')
plt.title('Trayectoria del balon',fontsize=14)
plt.xlabel('Distancia del tirador a la cesta', fontsize=14)
plt.ylabel('Coordenadas del balon en y', fontsize=14)
plt.legend()

```
