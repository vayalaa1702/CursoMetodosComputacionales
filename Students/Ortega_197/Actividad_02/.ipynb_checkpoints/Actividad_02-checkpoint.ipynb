#1. Escribir un programa en python en el que ingresada la profundidad máxima, cálcule punto a punto la variación de presión con la profundidad. Para ello el usuario deberá definir un valor máximo y en metros y el programa deberá retornar la grafica de Presión como función de y.

```python
X = np.linspace(0,30,100)
rho = 997 # [kg/m**3] densidad del agua
g = 9.8 # [m/s**2] aceleración de la gravedad
Pa = 101325 #Presión atmosferica, en pascales [Pa = kg/m*s**2]
f = lambda x: (Pa +rho*g*x)
F_=f(X)
plt.figure()
plt.title("Presión vs Profundidad")
plt.plot(X,F_,",-",label= "P(x)")
plt.xlabel("Profundidad (metros)",fontsize=10)
plt.ylabel("Presión (pascales)",fontsize=10)
plt.legend()
plt.show()
```

#2. Un jugador de baloncesto decide lanzar su balón a una distancia d del aro, con una rapidez v y ángulo theta respecto a la horizontal. Si la cesta tiene una altura de 3m, y el movimiento que realiza la pelota es un movimiento parabólico, construya un programa que determine si el jugador de baloncesto encesto o no encesto.

```python
#código de según un tetha,v0 y d dados encesta o no un baketbolista
import json
import numpy as np
import matplotlib.pylab as plt

def cesta(theta,vo,d):
  theta = (np.pi/180)*theta
  y = 1.60 #metros
  g = 9.8 #m/s**2
  h = 3 #metros
  t = d/(vo*np.cos(theta))
  time = np.linspace(0,t,100)


  y_t = y + vo*(np.sin(theta))*time - (g*time**2)/2

  y_f = y + vo*(np.sin(theta))*t - (g*t**2)/2
  #gráfica

  plt.plot(time,y_t)
  plt.plot(t,3,".")

  if abs(y_f - 3) < 0.01:
    return "Encestó"

  return "No encestó"

def cesta_json(data):
  vo = data["vo"]
  theta = data["theta"]
  d = data["d"]

  return cesta(theta,vo,d)
encesta = json.dumps({"theta": 45, "vo": 7.77, "d": 4})
with open("datos.json", "w") as f:
  f.write(encesta)

with open("datos.json") as f:
  data = json.load(f)
```

```python
cesta_json(data)
#datos para encestar
#angulo = 45°
#velocidad = 7.77
#altura = 1.60 metros
#distancia al aro = 4 metros
```
