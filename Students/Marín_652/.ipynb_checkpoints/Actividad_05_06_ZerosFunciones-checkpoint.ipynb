1. Implementar el metodo de la bisección para determinar las raices de una función.

Verficar para la función:

\begin{equation}
  f(x)=a e^{x}-2= b\cos(e^x-2)
\end{equation}

```python
def biseccion(f, xa, xb, tol=1e-6, Nmax=500):
  cont = 0
  while cont < Nmax:
    cont += 1
    p = (xb + xa) / 2  #punto medio
    if f(p) == 0 or abs(f(p)) < tol:
      return p
    elif f(p) * f(xa) < 0:  # Si f(p) y f(a) tienen signos opuestos, la raíz está en el intervalo [xa, p]
      xb = p
    else:  # Si no, la raíz está en el intervalo [p, xb]
      xa = p
  return p
```

```python
a=2
b=5
f=lambda x:a*np.exp(x) - 2 - b*np.cos(np.exp(x) - 2)
df=lambda x: a*np.exp(x)+b*np.sin(np.exp(x)-2)*np.exp(x)

x=np.linspace(-5,1.1,1000)
plt.figure(figsize=(10,6))
plt.plot(x,f(x),color = "crimson")
plt.xlabel("f(x)")
plt.ylabel("x")
plt.legend()
plt.title('Gráfica de f(x)=$a e^{x}-2= b\cos(e^x-2)$')
plt.grid(True)
plt.show()
```

```python
#Vemos que aproximadamente hay un cero en -3 y otro cero en 1
raiz = biseccion(f, -5, 1)
raiz1=biseccion(f, 0, 2)
print("La rimera raíz es :", raiz)
print("La segunda raíz es:", raiz1)
```

2. Implementar el metodo de la newton para determinar las raices de una función.

```python
#Métoodo de Newton-Rhapson
def newton_raphson(f, df, x0,tol=1e-6,Nmax=500):
  if abs(f(x0)) < tol:
    return print(f'Un cero de la funcion esta en {x0}')
  else:
    for i in range(Nmax):
      xn = x0 - f(x0)/df(x0)
      x0 = xn
      if abs(f(xn)) < tol:
        break
    if abs(f(xn)) > tol:
      return print('La funcion no tiene cero')
    return print(f'Un cero de la funcion esta en {x0}')

newton_raphson(f, df, 45)
```
