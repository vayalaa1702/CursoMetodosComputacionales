```python
def profundidad(h):
  g = 9.8  # m/s**2
  p = 1000  # kg/m**3
  y= np.zeros(h)
  P = np.zeros(h)
  for i in range (h):
    P[i]=p*g*(i+1)
    y[i]=i+1
  return y, P

h=15
y, P = profundidad(h)

plt.figure(figsize=(6, 4))
plt.plot(y, P, "o--", label="Función")
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Profundidad (metros)', fontsize=14)
plt.ylabel('Presión', fontsize=14)
plt.title("Variación de presión con la profundidad")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#FUNCION ENCESTÓ
def encesto(v0, theta, d, y0):
  thetar= np.deg2rad(theta)
  vx=v0*np.cos(thetar)
  v0y=v0*np.sin(thetar)
  tv=d/vx
  yf=y0+v0y*tv-(0.5*9.8*(tv**2))
  if abs(yf-3)<0.1:
    print('encesto')
  else:
    print('no encesto')
```

```python
def trayectoria(v0, theta, d, y0):
  thetar= np.deg2rad(theta)
  vx=v0*np.cos(thetar)
  v0y=v0*np.sin(thetar)
  tf=(v0y+((v0y**2)+2*(9.8)*y0)**(1/2))/9.8
  t=np.linspace(0,tf)
  x=-d+vx*t
  y=y0+v0y*t-(0.5*(9.8)*t**2)
  return x,y

def grafica_encesto(v0,theta, d, y0):
  encesto(v0,theta, d, y0)
  x,y = trayectoria(v0, theta, d, y0)
  fig, ax=plt.subplots(figsize=(10,5))
  ax.plot(x,y)
  plt.xlabel('Distancia (m)')
  plt.ylabel('Altura (m)')
  plt.title('Movimiento Parabólico del Lanzamiento de Baloncesto')
  ax.plot(0,3,'ro')


```

```python
import numpy as np
import matplotlib.pylab as plt
grafica_encesto(7,66,3,1.70)
```
