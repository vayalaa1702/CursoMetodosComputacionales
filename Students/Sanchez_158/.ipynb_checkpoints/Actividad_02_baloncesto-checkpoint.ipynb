```python
import json
import numpy as np
import matplotlib.pyplot as plt

def cargar_datos(nom_archivo):
    with open(nom_archivo, 'r') as archivo:
        datos = json.load(archivo)
    return datos

def encestar(datos):
    theta = np.radians(datos['theta'])
    vo = datos['vo']
    d = datos['d']
    g = 10 # m/s^2
    altura_canasta = 3

    #tiempo de vuelo
    t_vuelo = (vo * np.sin(theta) + np.sqrt((vo * np.sin(theta))**2 + 2 * g * altura_canasta)) / g

    # distancia horizontal recorrida en ese tiempo
    d_calculada = vo * np.cos(theta) * t_vuelo
    return t_vuelo, d_calculada

```

```python
print(encestar(datos))
```

```python
def grafica(datos):
    theta = np.radians(datos['theta'])
    vo = datos['vo']
    g = 10 # m/s^2

    # el tiempo de vuelo
    t_vuelo = (2 * vo * np.sin(theta)) / g

    # posiciones x y y en cada punto del tiempo
    tiempos = [i / 100 for i in range(int(t_vuelo * 100) + 1)]
    posiciones_x = [vo * np.cos(theta) * t for t in tiempos]
    posiciones_y = [vo * np.sin(theta) * t - 0.5 * g * t ** 2 for t in tiempos]

    # Grafica del movimiento parabólico
    plt.plot(posiciones_x, posiciones_y)
    plt.xlabel('Distancia Horizontal (m)')
    plt.ylabel('Altura (m)')
    plt.title('Movimiento Parabólico del Balón')
    plt.grid(True)
    plt.show()
```

```python
grafica(datos)
```
