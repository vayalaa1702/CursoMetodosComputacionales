Grafica de Presion a nivel del mar

```python
import numpy as np
import matplotlib.pyplot as plt
h = np.linspace(0,100,11)
#presion atmos, densidad del agua, gravedad, altura
Pa = 101325
G = 9.8
d = 997 # si es?
P = Pa + d*G*h
dif = P/Pa
for i in range(len(dif)):
    plt.title("Grafica de presión contra altura a nivel del mar")
    #plt.plot(h,-P,'.')
    plt.plot(h,-P,'-')
    plt.plot(h[i],-P[i],'o-',label= f'A una distancia de {h[i]}m se encuentra a una presion de {dif[i]}Pa')
    plt.xlabel("Metros sobre el nivel del mar",fontsize= 10)
    plt.ylabel("Presion sometida",fontsize= 10)
    plt.legend()
```

```python
import numpy as np

def tan(grados):
  return np.tan(np.deg2rad(grados))

def sen(grados):
  return np.sin(np.deg2rad(grados))

def cos(grados):
  return np.cos(np.deg2rad(grados))

def intan(grados):
  return np.rad2deg(np.arctan(grados))

def insen(grados):
  return np.rad2deg(np.arcsin(grados))

def incos(grados):
  return np.rad2deg(np.arccos(grados))

def cotan(grados):
  return 1/tan(grados)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
import matplotlib.animation as animation

Xo = 2
Yo = 2
g = 9.8
th_θ = 45
D = 10
H = 3
Vo = 5
T_dunk = (D-Xo)/(Vo*cos(th_θ))
T = np.linspace(0,T_dunk,1000)
X = Xo + (Vo*cos(th_θ)) * T
Y = Yo + (Vo*sen(th_θ)) * T - g/2 * (T**2)
Y = Y[Y>0]
X= np.sort(X[np.argsort(Y)])


fig, ax = plt.subplots()
scat = ax.scatter(T[0],Y[0], c="b",s=5,label=f'Vo = {Vo} m/s')
ax.set(xlim=[0,3],ylim=[-4,10],xlabel='Time [s]', ylabel='Y [m]')
ax.legend()

def update(frame):
    x = T[:frame]
    y = Y[:frame]
    # update the scatter plot:
    data = np.stack([x, y]).T
    scat.set_offsets(data)
    # update the line plot:
    return (scat)




ani = animation.FuncAnimation(fig=fig, func=update,frames=40,interval=30)
plt.show()
```
