Prueba tu resultado para:


  
 2. $a e^{x}-2= b\cos(e^x-2)$

 $0=-a e^{x}+ b\cos(e^x-2) +2$

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy as sc
from scipy import optimize
```

```python
x = np.linspace(-2,3,400)
fun = lambda x: -np.exp(x) + np.cos(np.exp(x)-2) + 2 # sin coeficientes
```

```python
y = fun(x)
root_bis = optimize.bisect(fun, 0,2)
plt.figure()
plt.plot(x, y)
plt.plot(root_bis,fun(root_bis), 'ro')
plt.grid()
plt.show()
```

```python
def func (x, a, b):
  return -a*np.exp(x) + b*np.cos(np.exp(x)-2) + 2

params  = (2,10)

root1_bisc = optimize.bisect(func,0,2,args=params)
root2_bisc = optimize.bisect(func,-1.5,-1,args=params)

print(root1_bisc,root2_bisc)

plt.figure()
plt.plot(x,func(x, *params), label = "Funcion")
plt.plot(root1_bisc, func(root1_bisc, *params), 'ro', label = "raices de la funcion")
plt.plot(root2_bisc, func(root2_bisc, *params),'ro')
plt.ylabel('$f(x) = -e^x + \cos(e^x - 2) + 2$')
plt.xlabel('x')
plt.legend()
plt.grid()
plt.show()
```

# METODO DE NEWTON_SECANTE

```python

params  = (2,10)

root1_sec = optimize.newton(func,2,args=params)
root2_sec = optimize.newton(func,-1,args=params)

print(root1_sec,root2_sec)

plt.figure()
plt.plot(x,func(x, *params), label = "Funcion")
plt.plot(root1_sec, func(root1_sec, *params), 'ro', label = "raices de la funcion")
plt.plot(root2_sec, func(root2_sec, *params),'ro')
plt.ylabel('$f(x) = -e^x + \cos(e^x - 2) + 2$')
plt.xlabel('x')
plt.legend()
plt.grid()
plt.show()

```
