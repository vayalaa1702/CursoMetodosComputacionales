```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import optimize
```

```python
a = 2
b = 5
funcion = lambda x: a * np.exp(x) - 2 - b * np.cos(np.exp(x) - 2)
error = 1e-5
Nmax = 500

def biseccion(f, xa, xb, epsilon, Nmax):
    niter = 0
    while niter < Nmax:
        xc = (xa + xb) / 2
        if abs(f(xc)) < epsilon:
            return xc
        elif f(xa) * f(xc) < 0:
            xb = xc
        else:
            xa = xc
        niter += 1
    print("No se encontró la raíz dentro del número máximo de iteraciones")
    return None

v = biseccion(funcion, -5, 1, error, Nmax)
print("El valor de la raíz es:", v)

x = np.linspace(-5, 1.5, 1000)
plt.figure()
plt.plot(x, funcion(x), "r--", color="#000000")
plt.xlabel("eje x")
plt.ylabel("eje y")
plt.plot(v,0,"o")
plt.title("grafica de la funcion")
plt.legend()
plt.show()
```
