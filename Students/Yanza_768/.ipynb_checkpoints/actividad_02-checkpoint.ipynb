Escribir un programa en python en el que ingresada la profundidad máxima, cálcule punto a punto la variación de presión con la profundidad. Para ello el usuario deberá definir un valor máximo y en metros y el programa deberá retornar la grafica de Presión como función de y. La gráfica deberá tener:
Unidades, titulo, fontsize=xticks=14
El label de la gráfica deberá indicar el numero de presiones atmosfericas a la que se esta sometida en produndidad máxima.

PROBLEMA DEL BUZO

```python
import numpy as np
import matplotlib.pyplot as plt
import math


#  El usuario escribe la profundidad maxima la que llega el buzo
profundidad_Max = float(input("Escribir la maxima profundidad en metros "))

# Y es mi dominio de funcion
Y = np.linspace(0,profundidad_Max,10)


# Funcion de Presion vs Profundidad Y. La presion es el rango. Uso  Sistema SI
Presion = lambda Y: 101325 + 1000*9.8*Y

# Evaluo la funcion en el arreglo Y divido por 101325 para obtener la presion en atmosferas
Presion_rango = Presion(Y)/101325

# Presión en atmosferas  en la profundidad maxima
P_maxima=Presion(profundidad_Max)/101325
```

```python
plt.figure(figsize=(6,4))
plt.plot(Y, Presion_rango, "o",label=f"# de P atmosfericas en  max profundidad ={P_maxima :.2f}")
plt.title(" Presión vs Profundidad ")
plt.xlabel("Profundidad [m] ",fontsize=15)
plt.ylabel("Presión [atm]",fontsize=15)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid()
plt.legend()
plt.show()
```

Un jugador de baloncesto decide lanzar su balón a una distancia d del aro, con una rapidez v y ángulo theta respecto a la horizontal. Si la cesta tiene una altura de 3m, y el movimiento que realiza la pelota es un movimiento parabólico, construya un programa que determine si el jugador de baloncesto encesto o no encesto. Para ello:

Definido theta, vo, y d como archivo json, se determine si el jugador encesto o no encesto.



```python
# IMPORTO EL MODULO JSON
import json

#Creo un diccionario o lista  con los datos que deseo  guardar en el archivo JSON.

datos = { "vO": 8, "theta_g": 60, "d": 4}

# Nombre del archivo JSON que deseo crear
baloncesto= 'datos.json'

# Guardo los datos en el archivo JSON
with open(baloncesto, 'w') as archivo:
    json.dump(datos, archivo)
```

```python
# Abrir JSON archivo
f = open('datos.json')

# retorna el archivo como un diccionario
data = json.load(f)


```

```python
def baloncesto(v,theta_g, d):
  y_cero = 2

  theta = math.radians(theta_g)

  y = y_cero + (d * np.tan(theta))- 0.5 * 9.8 * ((d/(v * np.cos(theta)))**2)

  vy = v * np.sin(theta) - 9.8 * (d/(v*np.cos(theta)))

  if y <= 3.2 and  y >= 2.8 and vy < 0:

    return "encestó !"
  else:
    return "no encestó !"

# los parametros son elementos del diccionario que importe del archivo json
cesta = baloncesto(data["vO"],data["theta_g"],data["d"])
print("\n  El jugador ", cesta)
```

Construir la gráfica del movimiento parabólico, para un theta, vo, y d, de tal forma que el jugador enceste

```python
vO = 7.27

angulo = math.radians(60)

di=4

t_vuelo= di / (vO * np.cos(angulo))

# el intervalo de tiempo va desde 0 hasta el tiempo de vuelo

T=np.linspace(0,t_vuelo, 10)

# POSICION HORIZONTAL
X = vO * np.cos(angulo) * T

# POSICION VERTICAL
Y =  2 + vO * np.sin(angulo) * T - 4.9 * (T**2)
```

```python
plt.figure(figsize=(6,4))
plt.plot(X, Y, "o",label="Trayectoria del balón")
plt.title(" Altura  vs Distancia horizontal ")
plt.xlabel(" Distancia horizontal [m] ",fontsize=15)
plt.ylabel("Altura [m]",fontsize=15)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid()
plt.legend()
plt.show()
```
