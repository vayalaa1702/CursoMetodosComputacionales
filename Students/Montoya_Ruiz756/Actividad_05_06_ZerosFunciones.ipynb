```python
import numpy as np
import math as mt
import matplotlib.pyplot as plt
from scipy import optimize
import pandas as pd
```

```python
def metodo_biseccion(f,a,b,n):
  error = 1e-6
  if (f(a)*f(b)) < 0:
    for i in range(n): # Numero de Iteracciones
      c = (a+b) *0.5
      if abs(f(c)) <= error: # Tolerancia de error
        return c
      elif f(c) > 0:
        b = c
      elif f(c) < 0:
        a = c
  else:
    print("No hay cero")
```

```python
def metodo_newton_rapshon(f,d,x_0,tol):
  iter = 0
  while iter < 100:
    if abs(f(x_0)) < tol:
      print(x_0)
      break
    else:
      x_0 = x_0 - f(x_0)/d(x_0)
    iter += 1
```

```python
def func(x,a,b):
  return (a*np.exp(x)) - 2 - (b*np.cos(np.exp(x)-2))
```

```python
a, b = 1,3
parametros = (a,b)
x = np.linspace(-5,5,1000)
f_ = func(x,a,b)
```

```python
d = {"x": x, "Funcion": f_}
df = pd.DataFrame(d)
df.plot.line(x = "x", y = "Funcion")
plt.grid()
plt.legend()
plt.show()
```

```python
df_filtro= df[df.Funcion.between(-0.1,0.1)]
df_filtro.plot.line(x = "x", y = "Funcion")
plt.grid()
plt.legend()
plt.show()
```

```python
raiz = optimize.bisect(func,1.150,1.154,args=parametros)
raiz
```

```python
raiz = optimize.newton(func,1.150,args=parametros)
raiz
```
