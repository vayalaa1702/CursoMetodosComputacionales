```python
# Librerias
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
# Punto 1 metodo de la biseccion
def biseccion(f, a, b):
  e = 1e-7
  if f(a)*f(b) > 0:
    return print('La funcion no tiene ceros en este intervalo')
  else:
    while f(a)*f(b) < 0:
      c = (a+b)/2
      if abs(f(c)) < e:
        return c
      elif f(a)*f(c) < 0:
        b = c
      elif f(b)*f(c) < 0:
        a = c
```

```python
# Punto 2 metodo de Newton-Raphson

def newton_raphson(f, df, x0):
  e = 1e-7
  if abs(f(x0)) < e:
    return x0

  else:
    for i in range(1000):
      xn = x0 - f(x0)/df(x0)
      x0 = xn
      if abs(f(xn)) < e:
        break

    if abs(f(xn)) > e:
      return print('La funcion no tiene cero')

    return x0
```

```python
# Deifnimos parametros de prueba
a = 0.5
b = 4.5

# Definimos nuestra funcion
def f(x):
  y = (a * np.exp(x)) - 2 - (b * np.cos(np.exp(x)-2))

  return y

x = np.linspace(-2,3, 10000)

# Graficamos la funcion para ver donde estan los ceros
plt.title('Grafica de f(x)')
plt.plot(x, f(x), label = 'f(x)')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

```python
# Probamos nuestros codigos para encontrar los ceros
print(f'Utilizando el metodo de la biseccion el cero es {biseccion(f, 1, 1.5)}')

# Calculamos la derivada de f para usar el metodo de Newton-Raphson
df = lambda x: a*np.exp(x) + b*np.sin(np.exp(x)-2)*np.exp(x)
print(f'Utilizando el metodo de Newton-Raphson el cero es {newton_raphson(f,df,1)}')
```
